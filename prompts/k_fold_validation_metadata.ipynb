```python
import os
import shutil
import pandas as pd

# Chemins à modifier selon ton cas
chemin_csv = "./validation.csv"
colonne_id = "file_name"  # nom de la colonne dans le CSV contenant les noms des images
dossier_source = "../VITON-HD-512/validation_images"
dossier_destination = "../VITON-HD-512/k_validations/"

# Lire le CSV
df = pd.read_csv(chemin_csv)

# S'assurer que le dossier destination existe
os.makedirs(dossier_destination, exist_ok=True)

# Boucler sur les IDs
for id_img in df[colonne_id]:
    # Si ce sont des IDs sans extension, ajoute une extension
    
    chemin_source = os.path.join(dossier_source, id_img)
    chemin_cible = os.path.join(dossier_destination, id_img)
    
    if os.path.exists(chemin_source):
        shutil.move(chemin_source, chemin_cible)  # ou shutil.copy() pour copier
        print(f"Déplacé : {id_img}")
    else:
        print(f"Fichier introuvable : {id_img}")
```

```python
for k in range(0,1):
    print(f"validation_{k}")
    
    import os
    import shutil
    import pandas as pd

    # Chemins à modifier selon ton cas
    chemin_csv = f"./validation_{k}.csv"
    colonne_id = "file_name" 
    dossier_source = "../VITON-HD-512/validation_images"
    dossier_destination = f"../VITON-HD-512/k_validations/validation_{k}"

    # Lire le CSV
    df = pd.read_csv(chemin_csv)

    # S'assurer que le dossier destination existe
    os.makedirs(dossier_destination, exist_ok=True)

    # Boucler sur les IDs
    for id_img in df[colonne_id]:
        # Si ce sont des IDs sans extension, ajoute une extension

        chemin_source = os.path.join(dossier_source, id_img)
        chemin_cible = os.path.join(dossier_destination, id_img)

        if os.path.exists(chemin_source):
            shutil.move(chemin_source, chemin_cible)  # ou shutil.copy() pour copier
            print(f"Déplacé : {id_img}")
        else:
            print(f"Fichier introuvable : {id_img}")


    df.to_csv(f"../VITON-HD-512/k_validations/validation_{k}/validation_{k}.csv", index=False)
```
