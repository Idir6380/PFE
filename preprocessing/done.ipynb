```python
import os
import shutil
import pandas as pd
```

```python
images_path = './group_outliers5/groups/unknown/'
csv_path = 'labels.csv'
done = './group_outliers5/groups/unknown/done'
```

```python
os.makedirs(done, exist_ok=True)
```

```python
df = pd.read_csv(csv_path)
```

```python
df
```

```python
for id in df['id']:
    img = f'{id}.jpg'
    src = os.path.join(images_path, img)
    dest = os.path.join(done, img)
    if os.path.exists(src):
        shutil.move(src, dest)
        print(f'{img} deplaced')
    else:
        print(f'{img} not preprocessed')
```
